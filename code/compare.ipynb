```python
import pandas as pd
```

## Adult Data Set

```python
adult = pd.read_csv("../data/adult.csv", header=None)
```

```python
adult['label'] = adult[14].apply(lambda x: 1 if x == " <=50K" else 0)
```

```python
adult
```

```python
adult = adult.dropna(axis=0, how='any')
data = adult.drop([14, 'label'], axis=1)
label = adult['label']
```

```python
data = pd.get_dummies(data)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data[[0,2,4,10,11,12]] = scaler.fit_transform(data[[0,2,4,10,11,12]])
```

```python
data
```

```python
train = data[0:380]
train_label = label[0:380]
test = data[380:]
test_label = label[380:]
```

## Cancer Data Set

```python
cancer = pd.read_csv("../data/cancer.csv", header=None)
```

```python
cancer[:10]
```

```python
goal = cancer[1]
```

```python
label = goal.apply(lambda x: 1 if x == "M" else 0)
```

```python
data = cancer.drop([0,1], axis=1)
```

```python
scaler = MinMaxScaler(copy=False)
scaler.fit_transform(data)
```

```python
data[:5]
```

```python
train = data[0:350]
train_label = label[0:350]
test = data[350:]
test_label = label[350:]
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import roc_auc_score
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(
    C=2.0,
)
lr.fit(train, train_label)
# lr.score(test, test_label)
```

```python
from xgboost.sklearn import XGBClassifier
xgb = XGBClassifier(
    learning_rate=0.02,
    subsample=0.5,
    n_estimators=3000,
    max_depth=3,
    reg_alpha=0.1,
    gamma=0.2
)
xgb.fit(train, train_label)
# xgb.score(test, test_label)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbdt = GradientBoostingClassifier(
    n_estimators=2000,
    max_depth=3,
    learning_rate=0.02,
    min_samples_split=3,
    subsample=0.5
)
gbdt.fit(train, train_label)
# gbdt.score(test, test_label)
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(
    hidden_layer_sizes=(55,),
    max_iter=1000,
    alpha=4
)
mlp.fit(train, train_label)
# mlp.score(test, test_label)
```

```python
model = mlp
pred = model.predict(test)
score = model.predict_proba(test)
score = [x[1] for x in score]
print(precision_score(test_label, pred))
print(roc_auc_score(test_label, score))
```

```python
model = mlp
pred = model.predict(train)
score = model.predict_proba(train)
score = [x[1] for x in score]
print(precision_score(train_label, pred))
print(roc_auc_score(train_label, score))
```

```python
wine = pd.read_csv("../data/wine.csv", delimiter=';')
wine[:5]
```

```python
target = wine['quality']
```

```python
data = wine.drop(['quality'], axis=1)
data[:5]
```

```python
scaler = MinMaxScaler(copy=False)
scaler.fit_transform(data)
```

```python
data[:5]
```

```python
train = data[:4200]
train_label = target[:4200]
test = data[4200:]
test_label = target[4200:]
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train, train_label)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbrt = GradientBoostingRegressor(
    n_estimators=1000,
    subsample=1,
    max_depth=3,
    min_samples_split=2,
    min_samples_leaf=3,
    learning_rate=0.01,
)
gbrt.fit(train, train_label)
```

```python
from xgboost.sklearn import XGBRegressor
xgb = XGBRegressor(
    n_estimators=1000,
    learning_rate=0.01,
    gamma=2,
    max_depth=3,
    reg_alpha=1,
    reg_lambda=1000,
    subsample=1
)
xgb.fit(train, train_label)
```

```python
from sklearn.neural_network import MLPRegressor
mlp = MLPRegressor(
    hidden_layer_sizes=(12,),
    max_iter=1000,
    alpha=30,
    learning_rate_init=0.1
)
mlp.fit(train, train_label)
```

```python
model = mlp
pred = model.predict(test)
print(mean_absolute_error(test_label, pred))
print(mean_squared_error(test_label, pred))
```

```python
model = mlp
pred = model.predict(train)
print(mean_absolute_error(train_label, pred))
print(mean_squared_error(train_label, pred))
```
