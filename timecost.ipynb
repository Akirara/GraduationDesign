```python
import pandas as pd
```

```python
adult = pd.read_csv("./data/adult.csv", header=None)
adult['label'] = adult[14].apply(lambda x: 1 if x == " <=50K" else 0)
adult = adult.dropna(axis=0, how='any')
data = adult.drop([14, 'label'], axis=1)
label = adult['label']
data = pd.get_dummies(data)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data[[0,2,4,10,11,12]] = scaler.fit_transform(data[[0,2,4,10,11,12]])
```

```python
data
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from xgboost.sklearn import XGBClassifier
```

```python
lr = LogisticRegression()
mlp = MLPClassifier(
    hidden_layer_sizes=(50, 20)
)
xgb = XGBClassifier(
    n_estimators=500
)
```

```python
lr.fit(data, label)
mlp.fit(data, label)
xgb.fit(data, label)
```

```python
test20 = data[:20]
test50 = data[:50]
test100 = data[:100]
```

```python
import time
```

```python
tmp = []
for i in range(10):
    tmp.append(test100)
test = pd.DataFrame()
test = test.append(tmp)

s = time.time()
lr.predict(test)
e = time.time()
print('lr:' + str(e-s))

s = time.time()
lr.predict(test)
mlp.predict(test)
xgb.predict(test)
e = time.time()
print('combine:' + str(e-s))

s = time.time()
lr.predict(test)
lr.predict(test20)
mlp.predict(test20)
xgb.predict(test20)
e = time.time()
print('K=20:' + str(e-s))

s = time.time()
lr.predict(test)
lr.predict(test50)
mlp.predict(test50)
xgb.predict(test50)
e = time.time()
print('K=50:' + str(e-s))

s = time.time()
lr.predict(test)
lr.predict(test100)
mlp.predict(test100)
xgb.predict(test100)
e = time.time()
print('K=100:' + str(e-s))
```
